```python
## pdf to img

import os
from PIL import Image
from pdf2image import convert_from_path

filePath = 'pdf/2020학년도_국어_홀.pdf'

doc = convert_from_path(filePath) # doc는 PIL.PpmImageFile list

# 파일 경로, 파일 이름(확장자포함) 구하기
path, fileName = os.path.split(filePath) # path, 파일name.pdf
print('path:', path, 'filename:', fileName)

for i in range(len(doc)):
    doc[i].save('ocr-image/2020-kor-'+str(i)+'.png') # 한글로 이름설정시 오류 issue

print('저장완료')
doc[0] # 1 page
```
